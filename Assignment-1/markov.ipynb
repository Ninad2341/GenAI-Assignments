```python
import random
from collections import defaultdict

# Training text
text = """
machine learning is a subset of artificial intelligence
artificial intelligence enables machines to learn from data
machine learning models improve with experience
"""
```

```python
# 1. Build Markov Chain (order-1)
markov_chain = defaultdict(list)

words = text.lower().split()
for i in range(len(words) - 1):
    markov_chain[words[i]].append(words[i + 1])
```

```python
# 2. Text generation function
def generate_text(start_word, length=15):
    word = start_word.lower()
    result = [word]

    for _ in range(length - 1):
        next_words = markov_chain.get(word)
        if not next_words:
            break
        word = random.choice(next_words)
        result.append(word)

    return " ".join(result)

# 3. Generate text
print(generate_text("machine", 20))
```
