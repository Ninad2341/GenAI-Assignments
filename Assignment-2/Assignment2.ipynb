```python
from groq import Groq

client = Groq(api_key="GROQ_API_KEY")

MODEL = "llama-3.3-70b-versatile"   
```

```python
def ask_llm(prompt):
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.7
    )
    
    return response.choices[0].message.content
```

```python
#Zero shot 
def zero_shot():
    prompt = "Classify the sentiment: 'This product is amazing!'"
    result = ask_llm(prompt)
    print("\nZero Shot Result:\n", result)
```

```python
#Few shot
def few_shot():
    prompt = """
    Classify sentiment:

    Text: I love this phone.
    Sentiment: Positive

    Text: This service is terrible.
    Sentiment: Negative

    Now classify:
    Text: The experience was fantastic.
    Sentiment:
    """
    
    result = ask_llm(prompt)
    print("\nFew Shot Result:\n", result)
```

```python
#chain of thought
def chain_of_thought():
    prompt = """
    Solve step-by-step:

    If a train moves at 60 km/h for 2.5 hours, what distance does it cover?
    Explain your reasoning clearly.
    """
    
    result = ask_llm(prompt)
    print("\nChain of Thought Result:\n", result)
```

```python
#Tree of thought
def tree_of_thought():
    prompt = """
    Problem: How can we increase user engagement in an AI Interview App?

    Step 1: Generate 3 different solution strategies.
    Step 2: Analyze pros and cons of each.
    Step 3: Select the best combined solution.
    """
    
    result = ask_llm(prompt)
    print("\nTree of Thought Result:\n", result)
```

```python
#Interview approach
def interview_mode():
    print("\nAI Interviewer Started\n")
    
    first_prompt = """
    You are a Data Science interviewer.
    Ask one Python question.
    """
    
    print("AI:", ask_llm(first_prompt))
    
    while True:
        user_input = input("Candidate: ")
        
        follow_prompt = f"""
        Candidate Answer: {user_input}
        Evaluate the answer briefly.
        Then ask a deeper follow-up question.
        """
        
        print("AI:", ask_llm(follow_prompt))
```

```python
if __name__ == "__main__":
    zero_shot()
    few_shot()
    chain_of_thought()
    tree_of_thought()
    
    
    # interview_mode()
```
